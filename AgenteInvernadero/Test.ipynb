```python
import os
import math
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence
```

```python
PATH = "../Models"
MODEL = os.path.join(PATH, "invernadero.bin")
```

```python
modelo = Word2Vec.load(MODEL)
wv = modelo.wv
```

```python
wv['zanahoria'][:30]
```

```python
print("Palabras más similares a 'zanahoria':")
similares_zanahoria = wv.most_similar(positive=['bajo'], topn=5)
for palabra, score in similares_zanahoria:
    print(f"  {palabra} → {score:.4f}")
```
